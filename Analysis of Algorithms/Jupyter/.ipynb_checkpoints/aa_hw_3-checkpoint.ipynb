<h3> QUESTION 1: IMPLEMENT MAX-CROSSING-SUBARRAY AND FIND-MAXIMUM-SUBARRAY <h3>

```python
import math

def Max_Crossing_Subarray(A, low, mid, high):
    left_sum = float("-inf")
    sum = 0
    max_left = 0
    max_right = 0
    for i in range(mid, low-1, -1):
        sum = sum + A[i]
        if sum > left_sum:
            left_sum = sum
            max_left = i
    right_sum = float("-inf")
    sum = 0
    for j in range(mid+1, high+1):
        sum = sum + A[j]
        if sum > right_sum:
            right_sum = sum
            max_right = j
    return(max_left, max_right, left_sum + right_sum)

def Find_Maximum_Subarray(A, low, high):
    if high == low:
        return (low, high, A[low])
    else:
        mid = math.floor((low+high)/2)
        (left_low, left_high, left_sum) = Find_Maximum_Subarray(A, low, mid)
        (right_low, right_high, right_sum) = Find_Maximum_Subarray(A, mid+1, high)
        (cross_low, cross_high, cross_sum) = Max_Crossing_Subarray(A, low, mid, high)
        if left_sum >= right_sum and left_sum >= cross_sum:
            return(left_low, left_high, left_sum)
        elif right_sum >= left_sum and right_sum >= cross_sum:
            return(right_low, right_high, right_sum)
        else:
            return(cross_low, cross_high, cross_sum)

print('Maximum subarray(start index, end index, value): '+ str(Find_Maximum_Subarray([-2,-3,4,-1,-2,1,5,-3], 0, 7)))
print('Maximum subarray(start index, end index, value): '+ str(Find_Maximum_Subarray([0,13,-3,-25,20,-3,-16,-23,18,20,-7,12,-5,-22,15,-4,7],0,16)))
```

When all values of A are negative, Find_Maximum_Subarray returns the biggest negative value in an array of length 1. The above code demonstrates this, the greatest possible subarray is a single number. There does not exist a subarray whose sum can be larger than ju, therefore -3 is the maximum subarray.

<h3> QUESTION 3: MAX PROFIT <h3>

```python
import csv

with open('AAPL.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    aapl = []
    high = []
    dateAAPL = []
    for row in csv_reader:
        if line_count == 0:
            line_count +=1
        elif line_count == 1:
            aapl.append(float(row[2]) - float(row[3]))
            high.append(float(row[2]))
            dateAAPL.append(row[0])
            line_count +=1
        else:
            highindex = line_count
            todayhigh = float(row[2])
            high.append(todayhigh)
            yesterdayhigh = high[highindex-2]
            difference = todayhigh - yesterdayhigh
            aapl.append(difference)
            dateAAPL.append(row[0])
            line_count += 1
    startIndex, endIndex, value = Find_Maximum_Subarray(aapl, 0, 250)
    print(dateAAPL[startIndex])
    print(dateAAPL[endIndex])
    print(Find_Maximum_Subarray(aapl, 0, 250))
```
